```python
import os
import cv2
import yt_dlp
import matplotlib.pyplot as plt


def baixar_video_youtube(url, output_path="video_atual.mp4"):
    """
    Baixa sempre um novo vídeo e sobrescreve corretamente.
    """
    # Remove arquivo antigo para evitar reutilizar vídeo anterior
    if os.path.exists(output_path):
        os.remove(output_path)

    ydl_opts = {
        'format': 'best[ext=mp4]/best',
        'outtmpl': output_path,
        'noplaylist': True,
        'quiet': False,
        'overwrites': True
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

    return output_path


def gerar_histogramas(video_path, intervalo_segundos=10, pasta_saida="histogramas"):
    """
    Lê o vídeo frame por frame e gera histogramas RGB.
    """
    os.makedirs(pasta_saida, exist_ok=True)

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"Erro ao abrir vídeo: {video_path}")
        return

    fps = cap.get(cv2.CAP_PROP_FPS)

    # Corrigir FPS inválido
    if fps <= 0:
        fps = 2

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duracao = total_frames / fps

    print(f"FPS: {fps:.2f}")
    print(f"Total de frames: {total_frames}")
    print(f"Duração: {duracao:.2f} segundos")

    frame_intervalo = max(1, int(fps * intervalo_segundos))

    frame_count = 0
    hist_count = 0

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        # Gera histograma em intervalos
        if frame_count % frame_intervalo == 0:

            if frame is None or frame.size == 0:
                frame_count += 1
                continue

            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            plt.figure(figsize=(14, 6))

            # Mostrar frame
            plt.subplot(1, 2, 1)
            plt.imshow(frame_rgb)
            tempo_atual = frame_count / fps
            plt.title(f"Frame em {tempo_atual:.2f}s")
            plt.axis("off")

            # Histograma RGB
            plt.subplot(1, 2, 2)

            for i, cor in enumerate(["red", "green", "blue"]):
                hist = cv2.calcHist([frame_rgb], [i], None, [256], [0, 256])
                plt.plot(hist, color=cor)

            plt.title("Histograma de Cores RGB")
            plt.xlabel("Intensidade")
            plt.ylabel("Pixels")
            plt.xlim([0, 256])

            nome_saida = os.path.join(
                pasta_saida,
                f"histograma_{hist_count:03d}_{int(tempo_atual)}s.png"
            )

            plt.tight_layout()
            plt.savefig(nome_saida, dpi=200)
            plt.close()

            print(f"Salvo: {nome_saida}")

            hist_count += 1

        frame_count += 1

    cap.release()

    print(f"\nProcesso concluído! {hist_count} histogramas gerados em '{pasta_saida}'")


youtube_url = "https://youtu.be/6dLEO8mwYWQ"
# ======================================
# CONFIGURAÇÃO
# ======================================
intervalo_segundos = 5

# ======================================
# EXECUÇÃO
# ======================================
video_path = baixar_video_youtube(youtube_url)

gerar_histogramas(
    video_path=video_path,
    intervalo_segundos=intervalo_segundos,
    pasta_saida="histogramas"
)
```

### Processamento de Múltiplos Vídeos com Limite de Histogramas

As funções `baixar_video_youtube` e `gerar_histogramas` foram atualizadas para permitir o processamento de uma lista de vídeos do YouTube e limitar a quantidade de histogramas gerados por vídeo.

```python
import os
import cv2
import yt_dlp
import matplotlib.pyplot as plt
import re # Para sanitizar nomes de arquivos/pastas
import random # Para selecionar frames aleatórios

def baixar_video_youtube(url, output_filename_prefix="video_baixado"):
    """
    Baixa um vídeo do YouTube. O nome do arquivo será prefixado com `output_filename_prefix`
    e um identificador sanitizado da URL.
    Retorna o caminho completo para o vídeo baixado.
    """
    # Sanitize URL para criar um identificador único no nome do arquivo
    sanitized_url_part = re.sub(r'[^a-zA-Z0-9]', '_', url)[:50] # Limita o tamanho
    output_path = os.path.join(
        os.path.dirname(output_filename_prefix), # Pega o diretório base
        f"{os.path.basename(output_filename_prefix)}_{sanitized_url_part}.mp4"
    )

    if os.path.exists(output_path):
        os.remove(output_path)

    ydl_opts = {
        'format': 'best[ext=mp4]/best',
        'outtmpl': output_path, # Usa o output_path dinâmico
        'noplaylist': True,
        'quiet': False,
        'overwrites': True,
        'verbose': False # Reduz a verbosidade do yt-dlp
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

    return output_path


def gerar_histogramas(video_path, intervalo_segundos=10, pasta_saida_base="histogramas_video", max_histograms_to_generate=2):
    """
    Lê o vídeo e gera histogramas RGB a partir de frames aleatórios, com um limite de histogramas.
    Os histogramas serão salvos em subpastas dentro de `pasta_saida_base`,
    nomeadas com base no vídeo processado.
    """
    # Cria uma pasta de saída única para os histogramas deste vídeo
    video_basename = os.path.basename(video_path).replace('.mp4', '')
    pasta_saida = os.path.join(pasta_saida_base, video_basename)
    os.makedirs(pasta_saida, exist_ok=True)

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"Erro ao abrir vídeo: {video_path}")
        return

    fps = cap.get(cv2.CAP_PROP_FPS)

    # Corrige FPS inválido ou zero, definindo um padrão.
    if fps <= 0:
        print(f"Aviso: FPS inválido ({fps}) para {video_path}. Usando FPS padrão de 30.")
        fps = 30

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duracao = total_frames / fps

    print(f"\n--- Processando vídeo: {video_path} ---")
    print(f"FPS: {fps:.2f}")
    print(f"Total de frames: {total_frames}")
    print(f"Duração: {duracao:.2f} segundos")

    # Seleciona N frames aleatórios
    frames_para_gerar = sorted(random.sample(range(total_frames), min(max_histograms_to_generate, total_frames)))

    hist_count = 0

    for frame_num in frames_para_gerar:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num) # Pula para o frame aleatório
        ret, frame = cap.read()

        if not ret:
            print(f"Aviso: Não foi possível ler o frame {frame_num}.")
            continue

        if frame is None or frame.size == 0:
            print(f"Aviso: Frame {frame_num} está vazio ou corrompido.")
            continue

        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        plt.figure(figsize=(14, 6))

        # Mostrar frame
        plt.subplot(1, 2, 1)
        plt.imshow(frame_rgb)
        tempo_atual = frame_num / fps
        plt.title(f"Frame em {tempo_atual:.2f}s (Frame #{frame_num})")
        plt.axis("off")

        # Histograma RGB
        plt.subplot(1, 2, 2)

        for i, cor in enumerate(["red", "green", "blue"]):
            hist = cv2.calcHist([frame_rgb], [i], None, [256], [0, 256])
            plt.plot(hist, color=cor)

        plt.title("Histograma de Cores RGB")
        plt.xlabel("Intensidade")
        plt.ylabel("Pixels")
        plt.xlim([0, 256])

        nome_saida = os.path.join(
            pasta_saida,
            f"histograma_{hist_count:03d}_{int(tempo_atual)}s.png"
        )

        plt.tight_layout()
        plt.savefig(nome_saida, dpi=200)
        plt.close()

        print(f"Salvo: {nome_saida}")

        hist_count += 1

    cap.release()
    print(f"Processo concluído para {video_path}! {hist_count} histogramas gerados em '{pasta_saida}'")
```

```python
# ======================================
# CONFIGURAÇÃO GLOBAL PARA MÚLTIPLOS VÍDEOS
# ======================================
youtube_urls = [
    "https://youtube.com/shorts/BtJBchqaUMI?si=KjhNpgV7Nt74dEk0", # Vídeo original
    "https://www.youtube.com/watch?v=Bl1FOKpFY2Q", # Exemplo 2
    "https://youtu.be/6dLEO8mwYWQ?si=hYb-0_YYTAbjFfoO",
    "https://youtu.be/iIWaMuQVa5g?si=qoq0HTERlRehwM7W",
    "https://youtu.be/7QpNACadX2s?si=0sVqN7Lwr-DKBboo",
    "https://youtu.be/uUQCzjx0ICI?si=ZZ96kcSMcJTP0gvT",
    "https://youtu.be/RndrcpjFaFE?si=9qQ8W15SUIx59yof",
    "https://youtu.be/lKvFe8c5btE?si=-JFQevEL1ASQwlLX",
    "https://youtu.be/2L0Cb18qTIs?si=ClaHZz3nLYCsIJEb",
    "https://youtu.be/riqEqoP0m_A?si=vmjdMpTLHEQHc2ZU",
    "https://youtu.be/TOiH9IgtAYQ?si=SeX1L42_fyT9Jjwl ",
    "https://youtu.be/h0caSK3aBuo?si=_hdfeSeRAYJltUzc"
    # Rick Astley - Never Gonna Give You Up
]
intervalo_segundos = 5
max_histograms_per_video = 3 # Limite de 2 histogramas por vídeo
base_output_dir = "videos_e_histogramas_multiplos" # Diretório base para todas as saídas

# Garante que o diretório de saída base exista
os.makedirs(base_output_dir, exist_ok=True)
```

```python
# ======================================
# EXECUÇÃO PARA MÚLTIPLOS VÍDEOS
# ======================================
for i, url in enumerate(youtube_urls):
    print(f"\n--- Iniciando processamento do Vídeo {i+1}/{len(youtube_urls)}: {url} ---")

    # Define caminhos únicos para cada vídeo
    video_output_filename_prefix = os.path.join(base_output_dir, f"video_{i+1}")
    histograms_output_folder_base = os.path.join(base_output_dir, "histogramas_gerados") # Todos os histogramas em uma pasta principal

    # Baixa o vídeo
    video_path = baixar_video_youtube(url, output_filename_prefix=video_output_filename_prefix)

    # Gera os histogramas
    gerar_histogramas(
        video_path=video_path,
        intervalo_segundos=intervalo_segundos,
        pasta_saida_base=histograms_output_folder_base,
        max_histograms_to_generate=max_histograms_per_video
    )

print("\n--- Todos os vídeos processados! ---")
```
